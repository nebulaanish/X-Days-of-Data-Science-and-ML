```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("Fifa 23 Fut Players.csv")
data.describe()
```

```python
data.head()
```

```python
x=np.array(data['Ratings'])
sns.distplot(x,bins=10)
# normal distribution as 
```

```python
print(data['Country'].nunique())
print(data['Country'].unique())
```

```python
high_rating = [data['Name'][i] for i in data['Ratings']]
high_rating_df=pd.DataFrame(high_rating,columns=['Name'])
high_rated_players = high_rating_df['Name'].unique()
high_rated_players
```

```python
#  Correlation between Ratings and popularity
print(data['Ratings'].corr(data['Popularity']))  # Low correlation
print(data['Ratings'].corr(data['DRI']))   # high correlation
```

```python
plt.figure(figsize=[20,10])
mask = np.triu(np.ones_like(data.corr(),dtype=bool))
sns.heatmap(data.corr(),mask=mask,cmap="Greens" ,annot=True)
```
